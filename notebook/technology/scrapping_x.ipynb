```python
import pymysql
from sqlalchemy import create_engine

# Konfigurasi koneksi database
db_user = "root"
db_password = "123456"
db_name = "db_hashscope"
db_ip = "127.0.0.1"
db_port = 3306

# Buat koneksi menggunakan pymysql
try:
    print("Membuka koneksi ke database...")
    connection = pymysql.connect(
        host=db_ip,
        user=db_user,
        password=db_password,
        database=db_name,
        port=db_port
    )
    print("Koneksi berhasil!")
except Exception as e:
    print("Gagal terhubung ke database:", e)
    raise SystemExit("Hentikan skrip karena koneksi gagal.")

# Data yang akan disisipkan
categories = [
    (1, "Technology"),
    (2, "Economy"),
    (3, "Politics")
]

# Query SQL untuk insert data
insert_query = "INSERT INTO main_categories (idmain_categories, category_name) VALUES (%s, %s);"

try:
    with connection.cursor() as cursor:
        print("Menyisipkan data ke tabel main_categories...")
        cursor.executemany(insert_query, categories)
        connection.commit()  # Simpan perubahan
        print("Data berhasil disisipkan.")
except Exception as e:
    print("Kesalahan saat menyisipkan data:", e)
    connection.rollback()  # Batalkan perubahan jika ada error
finally:
    connection.close()
    print("Koneksi ditutup.")
```

```python
import pandas as pd
from sqlalchemy import create_engine
import traceback
from datetime import datetime

# Konfigurasi koneksi database
db_user = "root"
db_password = "123456"
db_name = "db_hashscope"
db_ip = "127.0.0.1"
db_port = 3306

# Buat URL koneksi
db_connection_url = f"mysql+pymysql://{db_user}:{db_password}@{db_ip}:{db_port}/{db_name}"

# Buat engine SQLAlchemy
try:
    print("Menguji koneksi ke database...")
    engine = create_engine(db_connection_url)
    with engine.connect() as conn:
        print("Koneksi ke database berhasil!")
except Exception as e:
    print("Koneksi database gagal:")
    traceback.print_exc()
    raise SystemExit("Hentikan skrip karena koneksi database gagal.")

# Path file CSV
file_path = "tweets-data/x-technology.csv"

# Baca file CSV
try:
    print("Membaca file CSV...")
    df = pd.read_csv(file_path, delimiter=",")
    print(f"File CSV berhasil dibaca. Jumlah baris: {len(df)}")
except FileNotFoundError:
    raise SystemExit(f"File CSV tidak ditemukan di path: {file_path}")
except Exception as e:
    print("Kesalahan saat membaca file CSV:")
    traceback.print_exc()
    raise SystemExit("Hentikan skrip karena file CSV tidak dapat dibaca.")

# Pilih kolom yang relevan
try:
    print("Memilih kolom yang relevan...")
    df_selected = df[[
        "full_text", "favorite_count", "quote_count", "reply_count",
        "retweet_count", "username", "location", "created_at"
    ]].copy()

    # Tambahkan kolom idx_datasets
    print("Menambahkan kolom idx_datasets...")
    df_selected["idx_datasets"] = None

    # Tambahkan kolom main_categories_idmain_categories dengan nilai 1
    print("Menambahkan kolom 'main_categories_idmain_categories'...")
    df_selected["main_categories_idmain_categories"] = 1

    # Konversi format tanggal
    print("Mengonversi format kolom 'created_at'...")
    df_selected["created_at"] = pd.to_datetime(df_selected["created_at"], format="%a %b %d %H:%M:%S %z %Y").dt.strftime("%Y-%m-%d %H:%M:%S")
    print("Format kolom 'created_at' berhasil dikonversi.")

    # Urutkan kolom sesuai dengan tabel database
    df_selected = df_selected[[  # Urutan kolom harus sesuai dengan tabel di database
        "idx_datasets", "full_text", "favorite_count", "quote_count",
        "reply_count", "retweet_count", "username", "location", "created_at",
        "main_categories_idmain_categories"
    ]]
except KeyError as e:
    print("Kesalahan saat memilih kolom CSV:")
    traceback.print_exc()
    raise SystemExit("Hentikan skrip karena struktur file CSV tidak sesuai.")
except Exception as e:
    print("Kesalahan saat memproses data:")
    traceback.print_exc()
    raise SystemExit("Hentikan skrip karena kesalahan pada data.")

# Simpan data ke tabel database
try:
    print("Menyimpan data ke database...")
    df_selected.to_sql("x_datasets", con=engine, if_exists="append", index=False)
    print("Data berhasil disimpan ke Cloud SQL.")
except Exception as e:
    print("Kesalahan saat menyimpan data ke database:")
    traceback.print_exc()
```
